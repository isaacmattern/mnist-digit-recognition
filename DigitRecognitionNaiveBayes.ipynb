```python
from google.colab import drive
import pandas as pd
import math
import csv

drive.mount('/content/drive')
train = pd.read_csv('drive/My Drive/AI Final Project/train.csv')
test = pd.read_csv('drive/My Drive/AI Final Project/test.csv')
```

```python
num_counts = dict()
num_probs = dict()
total_count = 0

for index, row in train.iterrows():
  num_counts.update({row[0]: num_counts.get(row[0], 0) + 1})
  total_count += 1

for num in num_counts:
  num_probs.update({num: num_counts.get(num) / total_count})
```

```python
pixels = dict()
num_pixels = dict()
num_total_pixels = dict()

for num in num_counts:
  num_pixels[num] = dict()

for index, row in train.iterrows():
  pixel = 0
  for entry in row[1:]:
    pixel = pixel + 1
    if entry is 0:
      pixels.update({pixel: pixels.get(pixel, 0) + 1})
      num_pixels[row[0]].update({pixel: num_pixels[row[0]].get(pixel, 0) + 1})
      num_total_pixels.update({row[0]: num_total_pixels.get(row[0], 0) + 1})

total_pixels = len(pixels)
```

```python
alpha = 10
prob_is = dict()
pred_list = dict()

for index, row in test.iterrows():
  for num in num_counts:
    pixel = 0
    prob_is.update({num: math.log(num_probs[num])})
    for entry in row:
      pixel = pixel + 1
      numerator = alpha
      if entry is 0:
        if pixel in num_pixels[num]:
          numerator = numerator + num_pixels[num][pixel]
      denominator = num_total_pixels[num] + (alpha * total_pixels)
      prob_is.update({num: prob_is[num] + math.log(numerator / denominator)})
  pred_list.update({index + 1: max(prob_is, key = prob_is.get)})
```

```python
with open('drive/My Drive/AI Final Project/naive_bayes_results.csv', 'w') as f:
  w = csv.writer(f)
  for row in pred_list.items():
    w.writerow(row)
```
